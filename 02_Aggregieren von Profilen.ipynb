```python
#CSV selektieren

#OHNE Batterie:
#files = (glob.glob("data/data1/*.csv"))


#MIT Batterie:
files = (glob.glob("data/data2/*.csv"))

for file in files:
    print(file)
```

```python
#csv einlesen in df
df = None
for file in files:
    
    if df is None:
        df=pd.read_csv(file, error_bad_lines=False, low_memory=False, sep=";", decimal=",")
       # df.index = pd.to_datetime(df.index)
       # df
    
    else:
        df=df.append(pd.read_csv(file, error_bad_lines=False, low_memory=False, sep=";", decimal=","))
        #df.index = pd.to_datetime(df.index)
       

df.head(30)
```
